```python
countries = ['usa', 'denmark', 'india', 'brazil', 'russia']
exports = [68, 82, 78, 70, 76]
targets = [72, 78, 82, 76, 78]


index = np.arange(5)
w = 0.3

plt.figure(figsize=(8,8))
plt.bar(index, exports, w, color='green', label = "actual exports")
plt.bar(index, targets, w, color='red', label = "target exports", bottom=exports)

plt.title("exports data")
plt.xlabel("countries")
plt.ylabel("in millions")

plt.xticks(index, countries)
plt.legend(loc='best')

plt.show()
```

```python
countries = ['usa', 'denmark', 'india', 'brazil', 'russia']
exports = [68, 82, 78, 70, 76]
targets = [72, 78, 82, 76, 78]


index = np.arange(5)
w = 0.3

plt.figure(figsize=(8,8))
plt.bar(index, exports, w, color='green', label = "actual exports")
plt.bar(index, targets, w, color='red', label = "target exports", bottom=exports)

plt.title("exports data")
plt.xlabel("countries")
plt.ylabel("in millions")

plt.xticks(index, countries)
plt.legend(loc='upper right')

plt.show()
```
